System sterowania klimatyzacją, który reguluje temperaturę w pomieszczeniu na podstawie temperatury otoczenia i wilgotnośc

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

temperatura = ctrl.Antecedent(np.arange(0, 41, 1), 'temperatura')
wilgotnosc = ctrl.Antecedent(np.arange(0, 101, 1), 'wilgotnosc')
moc_klimatyzacji = ctrl.Consequent(np.arange(0, 101, 1), 'moc_klimatyzacji')

temperatura['zimno'] = fuzz.trimf(temperatura.universe, [0, 0, 20])
temperatura['umiarkowanie'] = fuzz.trimf(temperatura.universe, [10, 25, 35])
temperatura['goraco'] = fuzz.trimf(temperatura.universe, [25, 40, 40])

wilgotnosc['sucho'] = fuzz.trimf(wilgotnosc.universe, [0, 0, 50])
wilgotnosc['wilgotno'] = fuzz.trimf(wilgotnosc.universe, [50, 100, 100])

moc_klimatyzacji['slaba'] = fuzz.trimf(moc_klimatyzacji.universe, [0, 0, 50])
moc_klimatyzacji['srednia'] = fuzz.trimf(moc_klimatyzacji.universe, [25, 50, 75])
moc_klimatyzacji['silna'] = fuzz.trimf(moc_klimatyzacji.universe, [50, 100, 100])

# Definiowanie reguł rozmytych
regula1 = ctrl.Rule(temperatura['goraco'] & wilgotnosc['wilgotno'], moc_klimatyzacji['silna'])
regula2 = ctrl.Rule(temperatura['umiarkowanie'], moc_klimatyzacji['srednia'])
regula3 = ctrl.Rule(temperatura['zimno'] & wilgotnosc['sucho'], moc_klimatyzacji['slaba'])

system_sterowania = ctrl.ControlSystem([regula1, regula2, regula3])
symulacja = ctrl.ControlSystemSimulation(system_sterowania)

symulacja.input['temperatura'] = 30
symulacja.input['wilgotnosc'] = 70

symulacja.compute()
print(symulacja.output['moc_klimatyzacji'])
```

System sterowania suszarką do ubrań

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Definiowanie zmiennych lingwistycznych
wilgotnosc = ctrl.Antecedent(np.arange(0, 101, 1), 'wilgotnosc')
typ_tkaniny = ctrl.Antecedent(np.arange(0, 4, 1), 'typ_tkaniny')  # 0: delikatne, 1: bawelna, 2: jeans, 3: grube
czas_suszenia = ctrl.Consequent(np.arange(0, 61, 1), 'czas_suszenia')  # w minutach

# Definiowanie funkcji przynaleznosci
wilgotnosc['mokre'] = fuzz.trimf(wilgotnosc.universe, [60, 100, 100])
wilgotnosc['wilgotne'] = fuzz.trimf(wilgotnosc.universe, [30, 60, 80])
wilgotnosc['suche'] = fuzz.trimf(wilgotnosc.universe, [0, 20, 40])

typ_tkaniny['delikatne'] = fuzz.trimf(typ_tkaniny.universe, [0, 0, 1])
typ_tkaniny['bawelna'] = fuzz.trimf(typ_tkaniny.universe, [1, 1, 2])
typ_tkaniny['jeans'] = fuzz.trimf(typ_tkaniny.universe, [2, 2, 3])
typ_tkaniny['grube'] = fuzz.trimf(typ_tkaniny.universe, [3, 3, 3])

czas_suszenia['krotki'] = fuzz.trimf(czas_suszenia.universe, [0, 15, 30])
czas_suszenia['sredni'] = fuzz.trimf(czas_suszenia.universe, [20, 35, 50])
czas_suszenia['dlugi'] = fuzz.trimf(czas_suszenia.universe, [40, 50, 60])

# Definiowanie regul rozmytych
regula1 = ctrl.Rule(wilgotnosc['mokre'] & typ_tkaniny['grube'], czas_suszenia['dlugi'])
regula2 = ctrl.Rule(wilgotnosc['wilgotne'] & typ_tkaniny['bawelna'], czas_suszenia['sredni'])
regula3 = ctrl.Rule(wilgotnosc['suche'] & typ_tkaniny['delikatne'], czas_suszenia['krotki'])
regula4 = ctrl.Rule(wilgotnosc['mokre'] & typ_tkaniny['delikatne'], czas_suszenia['sredni'])

# System sterowania
system_sterowania = ctrl.ControlSystem([regula1, regula2, regula3, regula4])
symulacja = ctrl.ControlSystemSimulation(system_sterowania)

# Wartosci wejsciowe
symulacja.input['wilgotnosc'] = 70
symulacja.input['typ_tkaniny'] = 1  # bawelna

# Obliczenie wyniku
symulacja.compute()
print(symulacja.output['czas_suszenia'])
```

4.Zadania 1 System sterowania natężeniem światła w pokoju.


```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Definiowanie zmiennych lingwistycznych
pora_dnia = ctrl.Antecedent(np.arange(0, 24, 1), 'pora_dnia')  # godziny 0-23
preferencje = ctrl.Antecedent(np.arange(0, 11, 1), 'preferencje')  # 0-ciemno, 5-srednio, 10-jasno
natezenie_swiatla = ctrl.Consequent(np.arange(0, 101, 1), 'natezenie_swiatla')  # 0-100%

# Funkcje przynależności dla pory dnia
pora_dnia['rano'] = fuzz.trimf(pora_dnia.universe, [5, 8, 11])
pora_dnia['popoludnie'] = fuzz.trimf(pora_dnia.universe, [11, 14, 17])
pora_dnia['wieczor'] = fuzz.trimf(pora_dnia.universe, [17, 20, 23])

# Funkcje przynależności dla preferencji
preferencje['ciemno'] = fuzz.trimf(preferencje.universe, [0, 0, 5])
preferencje['srednio'] = fuzz.trimf(preferencje.universe, [3, 5, 7])
preferencje['jasno'] = fuzz.trimf(preferencje.universe, [5, 10, 10])

# Funkcje przynależności dla natężenia światła
natezenie_swiatla['slabe'] = fuzz.trimf(natezenie_swiatla.universe, [0, 0, 50])
natezenie_swiatla['srednie'] = fuzz.trimf(natezenie_swiatla.universe, [30, 50, 70])
natezenie_swiatla['silne'] = fuzz.trimf(natezenie_swiatla.universe, [50, 100, 100])

# Definiowanie reguł rozmytych
regula1 = ctrl.Rule(pora_dnia['rano'] & preferencje['jasno'], natezenie_swiatla['srednie'])
regula2 = ctrl.Rule(pora_dnia['rano'] & preferencje['ciemno'], natezenie_swiatla['slabe'])
regula3 = ctrl.Rule(pora_dnia['popoludnie'] & preferencje['jasno'], natezenie_swiatla['silne'])
regula4 = ctrl.Rule(pora_dnia['popoludnie'] & preferencje['srednio'], natezenie_swiatla['srednie'])
regula5 = ctrl.Rule(pora_dnia['wieczor'] & preferencje['ciemno'], natezenie_swiatla['slabe'])
regula6 = ctrl.Rule(pora_dnia['wieczor'] & preferencje['jasno'], natezenie_swiatla['srednie'])

# System sterowania
system_sterowania = ctrl.ControlSystem([regula1, regula2, regula3, regula4, regula5, regula6])
symulacja = ctrl.ControlSystemSimulation(system_sterowania)

# Testowanie systemu dla różnych wartości
# Przykład 1: rano i użytkownik chce jasno
symulacja.input['pora_dnia'] = 7
symulacja.input['preferencje'] = 9
symulacja.compute()
print(f"Rano, jasne preferencje: Natężenie światła = {symulacja.output['natezenie_swiatla']:.2f}%")

# Przykład 2: popołudnie i użytkownik chce średnio
symulacja.input['pora_dnia'] = 13
symulacja.input['preferencje'] = 5
symulacja.compute()
print(f"Popołudnie, średnie preferencje: Natężenie światła = {symulacja.output['natezenie_swiatla']:.2f}%")

# Przykład 3: wieczór i użytkownik chce ciemno
symulacja.input['pora_dnia'] = 21
symulacja.input['preferencje'] = 2
symulacja.compute()
print(f"Wieczór, ciemne preferencje: Natężenie światła = {symulacja.output['natezenie_swiatla']:.2f}%")
```

4.2 System oceny jakości powietrza

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Definiowanie zmiennych lingwistycznych
pm25 = ctrl.Antecedent(np.arange(0, 301, 1), 'pm25')  # 0-300 µg/m3
pm10 = ctrl.Antecedent(np.arange(0, 501, 1), 'pm10')  # 0-500 µg/m3
jakosc_powietrza = ctrl.Consequent(np.arange(0, 101, 1), 'jakosc_powietrza')  # 0-100% dobra

# Funkcje przynależności dla PM2.5
pm25['niskie'] = fuzz.trimf(pm25.universe, [0, 0, 75])
pm25['srednie'] = fuzz.trimf(pm25.universe, [50, 100, 150])
pm25['wysokie'] = fuzz.trimf(pm25.universe, [125, 300, 300])

# Funkcje przynależności dla PM10
pm10['niskie'] = fuzz.trimf(pm10.universe, [0, 0, 100])
pm10['srednie'] = fuzz.trimf(pm10.universe, [80, 150, 250])
pm10['wysokie'] = fuzz.trimf(pm10.universe, [200, 500, 500])

# Funkcje przynależności dla jakości powietrza
jakosc_powietrza['dobra'] = fuzz.trimf(jakosc_powietrza.universe, [60, 80, 100])
jakosc_powietrza['umiarkowana'] = fuzz.trimf(jakosc_powietrza.universe, [30, 50, 70])
jakosc_powietrza['zla'] = fuzz.trimf(jakosc_powietrza.universe, [0, 20, 40])

# Definiowanie reguł rozmytych
regula1 = ctrl.Rule(pm25['niskie'] & pm10['niskie'], jakosc_powietrza['dobra'])
regula2 = ctrl.Rule(pm25['srednie'] | pm10['srednie'], jakosc_powietrza['umiarkowana'])
regula3 = ctrl.Rule(pm25['wysokie'] | pm10['wysokie'], jakosc_powietrza['zla'])

# System sterowania
system_sterowania = ctrl.ControlSystem([regula1, regula2, regula3])
symulacja = ctrl.ControlSystemSimulation(system_sterowania)

# Testowanie systemu dla różnych wartości
# Przykład 1: Niskie PM2.5 i PM10
symulacja.input['pm25'] = 30
symulacja.input['pm10'] = 50
symulacja.compute()
print(f"Niskie PM2.5 i PM10: Jakość powietrza = {symulacja.output['jakosc_powietrza']:.2f}%")

# Przykład 2: Średnie PM2.5 i wysokie PM10
symulacja.input['pm25'] = 100
symulacja.input['pm10'] = 300
symulacja.compute()
print(f"Średnie PM2.5 i wysokie PM10: Jakość powietrza = {symulacja.output['jakosc_powietrza']:.2f}%")

# Przykład 3: Wysokie PM2.5 i niskie PM10
symulacja.input['pm25'] = 250
symulacja.input['pm10'] = 80
symulacja.compute()
print(f"Wysokie PM2.5 i niskie PM10: Jakość powietrza = {symulacja.output['jakosc_powietrza']:.2f}%")
```
